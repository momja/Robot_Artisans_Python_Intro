## Let's write some code
Try typing in the following code into a text editor or another Jupyter Notebook. 

```python
from PIL import Image

image = Image.open('image.jpg')
image.show()
```

- Make sure you have added an image called **image.jpg** to the directory you are in.
- In a Jupyter Notebook, execute by selecting the code and pressing `SHIFT+ENTER`. You might also want to try the code segment below if you want to load your image within the Jupyter notebook.
- If you are using another text editor, save the file with the .py extension. Then from a terminal, run `python3 --FILENAME--.py`

```python
from matplotlib.pyplot import imshow
import matplotlib.pyplot as plt
import numpy as np
from PIL import Image

%matplotlib inline # you can ignore this line for now
image = Image.open('image.jpg') # create variable image, and assign it to the image file provided
imshow(np.asarray(image)) # display the image in the notebook
```

From now on it will be assumed you are using a Juypter Notebook. 
Now let's try and manipulate our image:

```python
imshow(np.asarray(image.convert("1"))) # convert to 1 bit 2D array
```

The method `Image.convert(mode)` converts an RGB image to a bitmap. Read more about bitmaps here: http://paulbourke.net/dataformats/bitmaps/

There are also other modes you can convert to. Check the PIL (or Pillow) documentation for more info on image conversion: https://pillow.readthedocs.io/en/3.1.x/index.html


```python
for deg in range(0,180,20): # increase n by 20 each iteration, starting at 0 and ending at 180 (exclusive)
    plt.figure()
    imshow(np.asarray(image.rotate(deg)))
```
